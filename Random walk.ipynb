```python
import matplotlib.pyplot as plt
from random import choice
```

```python

class Randomwalks():
    def __init__(self, num_points = 5000):
        self.num_points = num_points
        
        # starting position
        self.x_values = [0]
        self.y_values = [0]
        
    def fill_walk(self):
        # Calculate all the point in the walk
        
        # loop that will keep taking steps until the walk reaches the desire length
        while len(self.x_values) < self.num_points:
            # Decide which direction to go and how far to go in that direction
            x_direction = choice([1,-1])
            x_distance = choice([0,1,2,3,4])
            x_step = x_direction * x_distance
            
            y_direction = choice([1,-1])
            y_distance = choice([0,1,2,3,4])
            y_step = y_direction * y_distance
            
            
            # reject moves that goes nowhere
            if x_step == 0 and y_step == 0:
                continue
            
            # calculate the next x and y values
            next_x = self.x_values[-1] + x_step
            next_y = self.y_values[-1] + y_step
            
            self.x_values.append(next_x)
            self.y_values.append(next_y)
```

```python
while True:
    rw = Randomwalks()
    rw.fill_walk()
    plt.scatter(rw.x_values, rw.y_values, s=5)
    plt.title('Randome scatter plot', fontsize = 20)
    plt.xlabel("Horizantal walk", fontsize= 14)
    plt.ylabel("Vertical walk", fontsize = 14)
    plt.tick_params(axis = 'both',labelsize = 10)
    plt.show()
    
    keep_running = input("Make another run (y/n): ")
    if keep_running == 'n':
        break
```
